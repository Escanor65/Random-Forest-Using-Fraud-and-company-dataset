## EDA

```python
data['SalesCategory'] = pd.cut(data['Sales'],bins=[0 , 5 , 10 , 15 , 20],labels=['Low Sales','Moderate Sales',
                                                                                 'High Sales','Very HighSales'])
```

```python
data1 = pd.get_dummies(data , columns=['ShelveLoc','Urban','US','SalesCategory'],drop_first= True) 
```

```python
data1
```

```python
X = data1.iloc[:,1:11]
```

```python
y = data1.iloc[:,12:15]
```

```python
y
```

## Random Forest Classifier
#### Criterion 'entropy'

```python
X_train , X_test , y_train , y_test = train_test_split(X, y, test_size=0.3 , random_state= 42)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
kfold = KFold(n_splits= 10 ,random_state= None)
model = RandomForestClassifier(n_estimators=100 , criterion= 'entropy' , max_features=3, min_samples_split=2,
    min_samples_leaf=1)
results = cross_val_score(model , X , y , cv = kfold )
```

```python
print('Accuracy :',results.mean()*100)
```

## Random forest Classifier using criterion 'gini'

```python
kfold = KFold(n_splits= 10 ,random_state= None)
model = RandomForestClassifier(n_estimators=100 , criterion= 'gini' , max_features=3, min_samples_split=2,
    min_samples_leaf=1)
results = cross_val_score(model , X , y , cv = kfold )
```

```python
print('Accuracy :',results.mean()*100)
```
